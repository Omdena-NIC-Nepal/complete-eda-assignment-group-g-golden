## Introduction
This notebook contains an exploratory data analysis of climate data from 1900 to 2023. The dataset includes global temperatures, CO2 concentration, sea level rise, and Arctic ice area.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot styling
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('viridis')
%matplotlib inline
```

## 1. Data Preparation

Load the climate data and perform necessary cleaning and aggregation.

```python
# Load the dataset
df = pd.read_csv('data/Climate_Change_Indicators.csv') # Place the correct path to the file you are reading here (Make sure to load using the relative path)

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values and basic information about the dataset
print("Dataset Information:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# TODO: Aggregate data by year to create a 124-year time series
#Create a data frame
df = pd.DataFrame(df)
print(df)

#Aggregated data

df = df.groupby('Year').mean().reset_index()

#Plot time series
#Global Average Temperature
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Global Average Temperature (°C)'], marker = 'o', color = 'r')
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Global Average Temperature per year (°C)")
plt.legend()
plt.show()

```

```python
#CO2 Concentration
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df['CO2 Concentration (ppm)'], marker = 'o', color = 'r')
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("CO2 Concentration per year(ppm)")
#plt.legend()
plt.grid(True)
plt.show()
```

```python
#Sea level rise
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Sea Level Rise (mm)'], marker = 'o', color = 'r')
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Sea Level Rise (mm) per Year")
plt.legend()
plt.grid(True)
plt.show()

```

```python
#Arctic Ice Area
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Arctic Ice Area (million km²)'], marker = 'o', color = 'r')
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Arctic Ice Area (million km² per Year)")
plt.legend()
plt.grid(True)
plt.show()
```
